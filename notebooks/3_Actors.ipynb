##### 10.3 concat & union

##### actors_concat1 & actors_union1 

```python
actors_concat1 = pd.concat([Actors_.Actor5,Actors_.Actor6,Actors_.Actor7,Actors_.Actor8], axis=1)
actors_concat1
```

```python
actors_concat1["actors_union1"] = (
    actors_concat1[['Actor5', 'Actor6', 'Actor7', 'Actor8']]
        .apply(lambda x: x.str.split(',\s*'))
        .sum(axis=1).map(np.array)
        .apply(lambda x: ', '.join(np.unique(x[x != 'empty'])))        
        .replace('', 'empty')
    )
actors_concat1.actors_union1
```

##### actors_concat2 & actors_union2

```python
actors_concat2 = pd.concat([movies_changes1.Actor1,movies_changes1.Actor2,movies_changes1.Actor3,movies_changes1.Actor4], axis=1)
actors_concat2
```

```python
actors_concat2.fillna('empty', inplace=True)
actors_concat2
```

```python
actors_concat2["actors_union2"] = (
    actors_concat2[['Actor1', 'Actor2', 'Actor3', 'Actor4']]
        .apply(lambda x: x.str.split(',\s*'))
        .sum(axis=1).map(np.array)
        .apply(lambda x: ', '.join(np.unique(x[x != 'empty'])))        
        .replace('', 'empty')
    )
actors_concat2.actors_union2
```

##### actors_concat3 & actors_union3

```python
actors_concat3 = pd.concat([movies_changes1.Star1,movies_changes1.Star2,movies_changes1.Star3,movies_changes1.Star4], axis=1)
actors_concat3
```

```python
actors_concat3.fillna('empty', inplace=True)
actors_concat3
```

```python
actors_concat3["actors_union3"] = (
    actors_concat3[['Star1', 'Star2', 'Star3', 'Star4']]
        .apply(lambda x: x.str.split(',\s*'))
        .sum(axis=1).map(np.array)
        .apply(lambda x: ', '.join(np.unique(x[x != 'empty'])))        
        .replace('', 'empty')
    )
actors_concat3.actors_union3
```

##### actors_concat4 & actors_union4 

```python
actors_concat4 = pd.concat([actors_concat1.actors_union1,actors_concat2.actors_union2, actors_concat3.actors_union3], axis=1)
actors_concat4
```

```python
actors_concat4["actors_union4"] = (
    actors_concat4[['actors_union1', 'actors_union2', 'actors_union3']]
        .apply(lambda x: x.str.split(',\s*'))
        .sum(axis=1).map(np.array)
        .apply(lambda x: ', '.join(np.unique(x[x != 'empty'])))        
        .replace('', 'empty')
    )
actors_concat4.actors_union4
```

#### 10.4 split actors_concat1.actors_union1 

```python
a_split = actors_concat4[['actors_union4']].copy()
a_split[['Actor_1', 'Actor_2', 'Actor_3', 'Actor_4']] = a_split['actors_union4'].str.split(',', expand=True)
a_split= a_split.drop(['actors_union4'], axis=1)
a_split
```

```python
a_split.Actor_1 = a_split['Actor_1'].str.replace('empty', '')
a_split.fillna('', inplace=True)
a_split
```

#### 10.5 removing the old columns and replcae with the a_split to movies_change1 (movies_change2)

```python
movies_changes1 = movies_changes1.drop(['Actor1', 'Actor2', 'Actor3', 'Actor4', 'Star1', 
                                        'Star2','Star3','Star4','Actors' ], axis=1)
movies_changes1.columns
```

```python
movies_changes2 = pd.concat([movies_changes1, a_split], axis=1)
movies_changes2.columns
```
