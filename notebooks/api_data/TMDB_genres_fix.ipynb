```python
import pandas as pd
import numpy as np
```

```python
tmdb_df1 = pd.read_csv("tmdb_df.csv")
```

```python
tmdb_df1.genres
```

```python
list_split = tmdb_df1[['genres']].copy()
list_split[['list1', 'list2', 'list3','list4', 'list5', 'list6', 'list7', 'list8', 'list9', 'list10', 'list11', 'list12', 'list13', 'list14', 'list15', 'list16', 'list17', 'list18']] = list_split['genres'].str.split(',', expand=True)
list_split= list_split.drop(['genres'], axis=1)
```

```python
list_split = list_split.drop(["list1","list3","list5","list7",
                              "list9","list11","list13","list15","list17"], axis=1)
list_split.head()
```

```python
list_split.columns
```

```python
list_split.columns
list_split.fillna('empty', inplace=True)
list_split.head()
```

```python
list_split["genres_fix"] = (
    list_split[['list2', 'list4', 'list6', 'list8', 'list10', 'list12', 'list14', 'list16']]
        .apply(lambda x: x.str.split(',\s*'))
        .sum(axis=1).map(np.array)
        .apply(lambda x: ', '.join(np.unique(x[x != 'empty'])))        
        .replace('', 'empty')
    )
list_split.genres_fix
```

```python
list_split.genres_fix = list_split.genres_fix.str.replace(" 'name': '","")
list_split.genres_fix = list_split.genres_fix.str.replace("'}]","")
list_split.genres_fix = list_split.genres_fix.str.replace("'}","")
list_split.genres_fix = list_split.genres_fix.str.replace(" ","")
```

```python
genres_tmdb = list_split[['genres_fix']].copy()
genres_tmdb[['genre1', 'genre2', 'genre3','genre4', 'genre5', 'genre6', 'genre7', 'genre8']] = genres_tmdb['genres_fix'].str.split(',', expand=True)
genres_tmdb= genres_tmdb.drop(['genres_fix'], axis=1)
genres_tmdb
```

```python
genres_tmdb.fillna('', inplace=True)
genres_tmdb
```

```python
genres_tmdb.to_csv('genres_tmdb.csv')
```

```python
tmdb_df1 = tmdb_df1.drop(['genres', 'Unnamed: 0'], axis=1)
tmdb_df1.columns
```

```python
tmdb_data = pd.concat([tmdb_df1, genres_tmdb], axis=1)
tmdb_data
```

```python
tmdb_data.to_csv('tmdb_data.csv')
```
